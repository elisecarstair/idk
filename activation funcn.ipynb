```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.maximum(0.1*x, x)

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

def binary_step(x):
    return np.heaviside(x, 1)

x = np.linspace(-5, 5, 100)

# Plot the sigmoid function
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()

# Plot the hyperbolic tangent function
plt.plot(x, tanh(x))
plt.title('Tanh Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()

# Plot the ReLU function
plt.plot(x, relu(x))
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()

# Plot the leaky ReLU function
plt.plot(x, leaky_relu(x))
plt.title('Leaky ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()

# Plot the binary function
plt.plot(x, binary_step(x))
plt.title('binary function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()

# Plot the softmax function
plt.plot(x, softmax(np.array([x, x*2, x*3]).T))
plt.title('Softmax Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

# Sure! An activation function is like a filter that processes the output of each neuron in a neural network.normalizes the output values
ReLU stands for Rectified Linear Unit and is one of the most popular activation functions used in neural networks. 
#The ReLU function takes an input value and returns either the input value itself (if it is positive), or zero (if it is negative)
ReLU introduces nonlinearity into the network, allowing it to learn more complex patterns in the data.


The sigmoid function takes an input value and "squashes" it into a value between 0 and 1.
f(x) = 1 / (1 + exp(-2)) = 0.880797
produces a smooth and differentiable output, which makes it easy to calculate the gradient for backpropagation during training. The sigmoid function is often used in the output layer of a binary classification problem, where the goal is to predict a binary outcome (e.g., true or false, yes or no). 


The tanh function takes an input value and "squashes" it into a value between -1 and 1.
f(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))
The tanh function is often used as an alternative to the sigmoid function in neural networks, particularly for classification tasks where the input values are normalized between -1 and 1.


multi-class classification problems
The softmax function takes a vector of input values and "squashes" them into a probability distribution over K different classes, where K is the number of possible classes.
f(x_i) = exp(x_i) / sum(exp(x_j))
Here's an example of how to compute the softmax function step-by-step for a vector of input values [1, 2, 3]:

Compute the exponential values for each element in the input vector:

exp(z) = [exp(1), exp(2), exp(3)]

Compute the sum of the exponential values:

sum(exp(z)) = exp(1) + exp(2) + exp(3)

Compute the output values for each element in the input vector using the softmax function formula:

y_1 = exp(1) / (exp(1) + exp(2) + exp(3)) = 0.090

y_2 = exp(2) / (exp(1) + exp(2) + exp(3)) = 0.245

y_3 = exp(3) / (exp(1) + exp(2) + exp(3)) = 0.665

Check that the output values sum up to 1:

y_1 + y_2 + y_3 = 0.090 + 0.245 + 0.665 = 1



The binary step function is a simple activation function that maps any input value to either 0 or 1.
f(x) = 0, if x < 0
       1, if x >= 0


The Leaky ReLU (Rectified Linear Unit) function is a modification of the ReLU function that overcomes its limitation of not being able to handle negative values well.
f(x) = x, if x >= 0
       alpha * x, if x < 0
Here, alpha is a small positive constant, usually set to a value between 0.01 and 0.5.
